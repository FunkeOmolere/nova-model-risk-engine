```python
import pandas as pd

def calculate_risk(complexity, data):
    score = (complexity * 2) + (data * 2)
    if score >= 16:
        tier = "High"
    elif score >= 10:
        tier = "Medium"
    else:
        tier = "Low"
    return score, tier

models = [
    {"name": "Credit Risk Model", "owner": "Ayo", "department": "Risk", "complexity": 5, "data": 5},
    {"name": "Marketing Model", "owner": "Zainab", "department": "Marketing", "complexity": 3, "data": 3},
    {"name": "Internal Forecast Model", "owner": "Tomi", "department": "Finance", "complexity": 2, "data": 2},
]

results = []
for m in models:
    score, tier = calculate_risk(m["complexity"], m["data"])
    results.append({
        "Model": m["name"],
        "Owner": m["owner"],
        "Department": m["department"],
        "Complexity": m["complexity"],
        "Data": m["data"],
        "Risk Score": score,
        "Tier": tier
    })

df = pd.DataFrame(results)
df
```

```python
import matplotlib.pyplot as plt

# Heatmap data: Owners (rows) x Departments (columns), showing Risk Score
pivot = df.pivot_table(
    index="Owner",
    columns="Department",
    values="Risk Score",
    aggfunc="max",   # or "mean"
    fill_value=0
)

plt.figure(figsize=(8, 4))
plt.title("Model Risk Heatmap (Risk Score)")

plt.imshow(pivot.values, aspect="auto")
plt.xticks(range(len(pivot.columns)), pivot.columns, rotation=45, ha="right")
plt.yticks(range(len(pivot.index)), pivot.index)

plt.colorbar(label="Risk Score")

# Put numbers on the heatmap
for i in range(pivot.shape[0]):
    for j in range(pivot.shape[1]):
        plt.text(j, i, int(pivot.values[i, j]), ha="center", va="center")

plt.tight_layout()
plt.show()
```

```python
def calculate_risk(complexity, data_sensitivity, regulatory_exposure):
    score = (
        complexity * 0.4 +
        data_sensitivity * 0.35 +
        regulatory_exposure * 0.25
    ) * 20  # Scale to 100
    
    if score >= 70:
        tier = "High"
    elif score >= 40:
        tier = "Medium"
    else:
        tier = "Low"
        
    return round(score, 2), tier
```

## Risk Scoring Logic

We calculate risk using three components:
- Model complexity
- Data sensitivity
- Regulatory exposure

Each factor is scored 1–5 and weighted to produce a total risk score.
